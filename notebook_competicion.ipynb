#CARGA DEL MODELO FINAL

Lo primero, vamos a importar las librerías para cargar el modelo del otro cuaderno, y Pandas para leer el CSV.

```python
from joblib import load
import pandas as pd
```

Carga del modelo final: Ridge




```python
modelo_final = load('modelo_final.pkl')
```

Lectura de los datos

```python
comp_df = pd.read_csv("comp_st4ns1.txt.bz2",
 compression="bz2",
 index_col=0)

comp_df
```

Cogemos los datos de las 75 variables por un lado, separándolas del atributo de salida

```python
X = comp_df.iloc[:, 0:75]
```

Predicción con X utilizando el modelo final, y generación de los datos de salida pedidos

```python
predicciones_competicion = modelo_final.predict(X)

predicciones_competicion
```

Importación a CSV de los datos de salida y generación del fichero

```python
df = pd.DataFrame(predicciones_competicion, columns=['predictions'])

df.to_csv('predicciones.csv', index=False)
```
